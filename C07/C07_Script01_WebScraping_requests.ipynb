# Wikipedia

```python
# Definamos la URL semilla a scrapear
url_wiki = "https://wikipedia.org/"
```

```python
#Obtengamos la "información del servidor": Requerimieto
respuesta = requests.get(url = url_wiki)
respuesta

# 403 Forbidden
# El cliente no posee los permisos necesarios para cierto contenido, por lo que
# el servidor está rechazando otorgar una respuesta apropiada
```

```python
# Vamos a configurar un user-agent para que el servidor Wikipedia no nos banee
# Para ello hay que implementar un script que seleccione de manera aleatoria
# una fila del siguiente archivo:
# https://gist.githubusercontent.com/pzb/b4b6f57144aea7827ae4/raw/cf847b76a142955b1410c8bcef3aabe221a63db1/user-agents.txt

# Descarguemos el archivo user-agent
!wget https://gist.githubusercontent.com/pzb/b4b6f57144aea7827ae4/raw/cf847b76a142955b1410c8bcef3aabe221a63db1/user-agents.txt
```

```python
#
with open("/content/user-agents.txt", "r") as f:
    lines = f.readlines()

lines
```

```python
lines[0]
```

```python
import random as rnd
n = len(lines)
headers = {"user-agent": lines[rnd.randint(0,n+1)][:-1]}

respuesta = requests.get(url = url_wiki, headers = headers)
respuesta
```

```python
# Podemos notar que del arbol del HTML que los id del número de artículos de cada
# idioma están en una clase "central-featured-lang"
idiomas = parser.find_class("central-featured-lang")
idiomas
```

```python
idiomas[0]
```

```python
idiomas[0].text_content() # Inglés
```

```python
idiomas[1].text_content() # Chino
```

```python
idiomas[2].text_content().split()
```

```python
idiomas[2].text_content().split()[1:]
```

```python
# Construyamos una lista con los idiomas y una lista para el número aproximado de artículos

lista_idiomas = []
lista_num_aprox_articulos = []

# Usemos una estructura repetitiva para barrer todos los elementos del iterable idioma
for idioma in idiomas:
    lista_idiomas.append(idioma.text_content().split()[0])
    lista_num_aprox_articulos.append(idioma.text_content().split()[1:])

print("Lista de idiomas: ", lista_idiomas)
print("Lista de números aproximados de artículos: ", lista_num_aprox_articulos)
```
