```python
from matplotlib import lines
# Crear el histograma sin el kde y luego agregar una línea kde

# Podemos crear un contendendor (FIgure) al Axes donde colocaremos el histograma
# y la curva KDE
plt.figure(figsize = (16,9))

# Capa 1: Histograma (Solo barritas, son KDE)
ax = sns.histplot(
    data=tips["tip"],
    bins = 15,
    stat = "density",
    kde = False,
    # bw_adjust < 1 → La curva se pega más a los datos: Más ruido/varianza, menos sesgo
    # bw_adjust > 1 → La curva es más suave/generalista: Menos ruido/varianza, más sesgo
    kde_kws={"bw_adjust":0.5},
    color = "orange",
    label = " Estos son los datos observados (Histograma)"
    )

# Capa 2: Agreguemos la curva kde
sns.kdeplot(
    data = tips["tip"],
    color = "green",
    linewidth = 1.5,
    linestyle = "--",
    label = " Curva de densidad (KDE)",
    ax = ax
)

# Personalizar la leyenda
plt.legend(
    title_fontsize = 6,
    frameon = False
)

# Agrego el método legend
plt.legend()
```

```python
# Estrategia 1: Estadísticas de propinas por género

# Creamos el Figure y la matriz de Axes
fig, axes = plt.subplots(2, 2, figsize=(8, 8))

# Cuadrante superior izquierdo: axes[0,0] (Mediana)
sns.barplot(
    data = tips,
    x = "sex",
    y = "tip",
    hue = "sex",
    palette=["red", "blue"],
    estimator = np.mean,
    ax = axes[0,0],
    errorbar = None
)

# A este axes[0,0] debo de colocarle sus indicadores (media)
for container in axes[0,0].containers:
    axes[0,0].bar_label(
        container = container,
        color = "white",
        label_type = "center",
        fmt = "%.2f",
        fontsize = 25,
        fontweight = "bold"
        )
axes[0,0].set_title("Media")

# De igual manera se crea el barplot y se agrega las etiquetas a cada barra de cada axes
# Cuadrante inferior derecho: Mínimo axes[1,1] (Mínimo)
sns.barplot(
    data = tips,
    x = "sex",
    y = "tip",
    hue = "sex",
    palette=["red", "blue"],
    estimator = np.min,
    ax = axes[1,1],
    errorbar = None
)

# A este axes[1,1] debo de colocarle sus indicadores (mínimo)
for container in axes[1,1].containers:
    axes[1,1].bar_label(
        container = container,
        color = "white",
        label_type = "center",
        fmt = "%.2f",
        fontsize = 25,
        fontweight = "bold"
        )
axes[1,1].set_title("Mínimo")

# Cuadrante superior izquierdo: axes[0,1] (Mediana)
sns.barplot(
    data = tips,
    x = "sex",
    y = "tip",
    hue = "sex",
    palette=["red", "blue"],
    estimator = np.median,
    ax = axes[0,1],
    errorbar = None
)

# A este axes[0,1] debo de colocarle sus indicadores (mediana)
for container in axes[0,1].containers:
    axes[0,1].bar_label(
        container = container,
        color = "white",
        label_type = "center",
        fmt = "%.2f",
        fontsize = 25,
        fontweight = "bold"
        )
axes[0,1].set_title("Mediana")

# Cuadrante inferior izquierdp: Máximo axes[1,0] (Máximo)
sns.barplot(
    data = tips,
    x = "sex",
    y = "tip",
    hue = "sex",
    palette=["red", "blue"],
    estimator = np.max,
    ax = axes[1,0],
    errorbar = None
)

# A este axes[1,0] debo de colocarle sus indicadores (mínimo)
for container in axes[1,0].containers:
    axes[1,0].bar_label(
        container = container,
        color = "white",
        label_type = "center",
        fmt = "%.2f",
        fontsize = 25,
        fontweight = "bold"
        )
axes[1,0].set_title("Máximo")

# Fin
plt.tight_layout()
```

```python
# Estrategia 2:

# Fase 1
def calcular_estadisticos(grupo):
    return pd.Series(
        {
            "Mediana": grupo.median(),
            "Media": grupo.mean(),
            "Maximo": grupo.max(),
            "Minimo": grupo.min()
        }
    )

# Fase 2
stats_df = tips.groupby("sex", observed=False)["tip"].apply(calcular_estadisticos).unstack()

# Fase 3: Remodelado de datos (wide to long transformation)
stats_melt = stats_df.reset_index().melt(
    id_vars = "sex",
    value_vars = ["Mediana", "Media", "Maximo", "Minimo"], # Columna despivotar
    var_name = "Estadistico",
    value_name = "Valor"
)

# Fase 4
plt.figure(figsize = (8, 8))

ax = sns.barplot(
    data = stats_melt,
    x = "sex",
    y = "Valor",
    hue = "Estadistico",
    palette = "Set2")

# Iterar sobre los contenedores
for container in ax.containers:
    ax.bar_label(
        container = container,
        color = "white",
        label_type = "center",
        fmt = "%.2f",
        fontsize = 12,
        fontweight = "bold"
    )
```
