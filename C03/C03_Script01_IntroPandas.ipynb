# El Módulo Pandas
  1. Datos de tipo Series y DataFrame
  2. Indexación y selección

```python
# Carguemos los módulos
import pandas as pd
import numpy as np
```

```python
# Los tipos de datos que provee pandas de manera nativa son los Series (Unidimensional)
# y los DataFrames(Bidimensionales)
```

```python
# Veamos cómo crear un objeto unidimensional de pandas: pandas.Series

# Consideremos construir un dato de tipo list
l1 = [12, 234, 4,6,0,1,2,3,4]
s1 = pd.Series(l1)
s1
```

```python
# Considemos construir un dato de ndarray (propio del módulo numpy)
vector1 = np.array(l1)
s2 = pd.Series(vector1)
s2
```

```python
# También podemos crear un dato de tipo Series a partir de un dato de tipo diccionario
d1 = {"a1": 12.3,"a2":-5.6, "a3":9.81, "a4":10}
s3 = pd.Series(d1)
s3
```

```python
# También podemos crear un Series con índices en particular si utilizamos dos argumentos en pd.Series
s4 = pd.Series(data = [12,34,45,6,7,8,0,1], index=["a1","a2","a3","a4","a5", "a6","a7", "a8"])
s4
```

```python
s5 = pd.Series(np.random.triangular(2,3.5,6, 120))
s5
```

# Otro dataset
Es el listado de pacientes a los que le realizarón una cirugía de alta complejidad en el Instituto Nacional de Enfermedades Neoplásicas - INEN, cada registro corresponde a un paciente, donde se tiene  historia, sexo, fecha de operación , hora de inicio de la cirugia, hora que finaliza la cirugía, código de ubigeo del lugar de residencia  y  lugar residencia.

Link: `https://datosabiertos.gob.pe/dataset/listado-de-cirug%C3%ADas-de-alta-complejidad-realizadas-de-enero-2022-julio-2025-%E2%80%93-instituto-0`

image.png

```python
cirugias = pd.read_csv("https://datosabiertos.gob.pe/sites/default/files/Listado_cirugia_alta_complejidad_realizadas_ene2022_jul2025.csv",
                       # Agregamos el argumento encoding para resolver el siguente problema
                       # UnicodeDecodeError: 'utf-8' codec can't decode byte 0xd1 in position 948: invalid continuation byte
                       encoding = "ISO-8859-1"
                       )
cirugias.info()
```

```python
# Analicemos algo de información de la columna SEXO
cirugias["SEXO"]
```

```python
# Encontremos todos los valores únicos que componen a la columna SEXO
cirugias["SEXO"].unique()
```

```python
# Contemos la frecuencia de estos posibles dos valores de la columna SEXO
cirugias["SEXO"].value_counts()
```

```python
# En porcentaje
cirugias["SEXO"].value_counts()*100/cirugias.shape[0]
```

```python
# Analicemos a la columna FECHA_OPERACION: Almacena información temporal. Además podemos
# notar que no sigue unformato estandar
cirugias["FECHA_OPERACION"]
```

```python
# Vamos a crear nuevas columnas+
# Creación de la columna AÑO_OPERACION: Formada por los primeros 4 caracteres
cirugias["AÑO_OPERACION"] = cirugias["FECHA_OPERACION"].astype(str).str[:4].astype(int)
cirugias["MES_OPERACION"] = cirugias["FECHA_OPERACION"].astype(str).str[4:6].astype(int)
cirugias["DIA_OPERACION"] = cirugias["FECHA_OPERACION"].astype(str).str[6:8].astype(int)
cirugias
```

```python
# Contar el número de operaciones realizadas por año
cirugias["AÑO_OPERACION"].value_counts().sort_index()
```

```python
# El número de operaciones por mes y por cada año
mes_2022 = cirugias.loc[cirugias.AÑO_OPERACION == 2022, "MES_OPERACION"].value_counts().sort_index()
mes_2023 = cirugias.loc[cirugias.AÑO_OPERACION == 2023, "MES_OPERACION"].value_counts().sort_index()
mes_2024 = cirugias.loc[cirugias.AÑO_OPERACION == 2024, "MES_OPERACION"].value_counts().sort_index()

mes_operacion = pd.concat([mes_2022, mes_2023, mes_2024], axis=1, keys=["2022", "2023", "2024"])
mes_operacion
```

```python
# Crucemos la información que provee la columna SEXO y la columna AÑO_OPERACION
pd.crosstab(cirugias["SEXO"], cirugias["AÑO_OPERACION"])
```

# Carguemos otro dataset

Link: `https://www.mef.gob.pe/es/?option=com_content&language=es-ES&Itemid=102666&lang=es-ES&view=article&id=6430`

```python
# Carguemos el dataset del programa reactiva perú usando pandas para procesar cierta información
rctva = pd.read_excel("/content/Reactiva_Peru_Lista_de_empresas_al_30102020.xlsx",
                     # Luego de abrir el archivo xlsx notamos que debemos olvidar a las
                      # primeras dos filas; además también notamos que debemos olvidar a las
                      # últimas tres filas.
                      skiprows=2,
                      skipfooter=3
                      )
rctva.info()
```

```python
# Podemos notar que algunas empresas obtuvieron son un crédito, y algunas de estas obtuvieron dos créditos

# Las empresas que solo obtuvieron un crédito, son aquellas cuyo valor en la columna
# NOMBRE DE 2DA. ENTIDAD OTORGANTE DEL CRÉDITO* es NaN (Not a Number)
rctva_1cred = rctva.loc[rctva["NOMBRE DE 2DA. ENTIDAD OTORGANTE DEL CRÉDITO*"].isna(), :]

# Empresas que obtuvieron los dos créditos
rctva_2cred = rctva.loc[~rctva["NOMBRE DE 2DA. ENTIDAD OTORGANTE DEL CRÉDITO*"].isna(), :]
```

```python
# Analicemos el "SECTOR ECONÓMICO" de las empresas que obtuvieron solo un crédito
sec_eco_abs = rctva_1cred["SECTOR ECONÓMICO"].value_counts()
sec_eco_rel = rctva_1cred["SECTOR ECONÓMICO"].value_counts()*100/rctva_1cred.shape[0]

pd.concat([sec_eco_abs, sec_eco_rel], axis=1, keys=["Absoluto", "Relativo"])
```

```python
# Analicemos el "SECTOR ECONÓMICO" de las empresas que obtuvieron dos créditos
sec_eco_abs = rctva_2cred["SECTOR ECONÓMICO"].value_counts()
sec_eco_rel = rctva_2cred["SECTOR ECONÓMICO"].value_counts()*100/rctva_2cred.shape[0]

pd.concat([sec_eco_abs, sec_eco_rel], axis=1, keys=["Absoluto", "Relativo"])
```

```python
# Veamos qué es lo que ocurre a nivel de departamento
rctva_1cred["DEPARTAMENTO"].value_counts()
```

```python
# Para las empresas que obtuvieron 1 crédito, ma media de "MONTO PRÉSTAMO (s/)" por sector
# económico
rctva_1cred[["SECTOR ECONÓMICO","MONTO PRÉSTAMO (S/)"]]

# Inciemos por un diccionario en blanco
dict_sec_eco = {}.fromkeys(rctva_1cred["SECTOR ECONÓMICO"].unique())
for sec_eco in rctva_1cred["SECTOR ECONÓMICO"].unique():
  dict_sec_eco[sec_eco] = float(rctva_1cred.loc[rctva_1cred["SECTOR ECONÓMICO"] == sec_eco, "MONTO PRÉSTAMO (S/)"].sum())

pd.DataFrame(dict_sec_eco, index=["suma_monto_prestamo"]).T.sort_values(by="suma_monto_prestamo", ascending=False)
```
