# Analizar las columnas espaciales: `lat` y `lon`

```python
# Verificar el rango geográfico de los datos
# Longitud: -119 a -118
# Latitud: 33 a 35

# Construir índices que nos permitan validar la pertenencia de cada observación (fila)
valid_lat = (crime_la["lat"] >= 33) & (crime_la["lat"] <= 35)
valid_lon = (crime_la["lon"] >= -119) & (crime_la["lon"] <= -118)

# Construir un objeto (dataframe) cuyas columnas lat y lon están validadas
valid_geo = crime_la[valid_lat & valid_lon]

# Número de coordenadas inválidas
print(f"Número de coordenadas inválidas: {len(crime_la) - len(valid_geo)}")
```

```python
# Mapa de calor interactivo: folium

import folium
from folium.plugins import HeatMap

# Primera buena práctica: Seleccionar una muestra de los datos totales (> 10^6)
sample_size = 50000
sample_geo = valid_geo.sample(n = sample_size)

# Debemos juntar latitud y longitud en una lista: [lat, lon]
# Esto provee la data para el mapa de calor (HeatMap)
heat_data = [[row["lat"], row["lon"]] for indices, row in sample_geo.iterrows()]

# Creamos un mapa centrado en los Ángeles
la = folium.Map(location = [34.0522, -118.2437], zoom_start = 10)

# Agregamos una capa encima de la anterior para el mapa de calor
HeatMap(
  data = heat_data,
  radius = 10,
  min_opacity = 0.3,
  max_zoom = 18
).add_to(la)

# Mostramos el Mapa
la
```

```python
# https://www.senamhi.gob.pe/site/descarga-datos/
# Estación: Santa Rosa

# Leer el archivo de datos y procesar su contenido
def procesar_archivo(ruta):
    # Bloque try
    try:
        with open(ruta, "r") as f:
            lineas = f.readlines()

        datos = []
        lineas_con_error = 0

        for i, linea in enumerate(lineas, start=1):
            linea = linea.strip()
            if not linea:
                continue  # Ignoramos las líneas en blanco

            # Dividir por espacios en blanco
            campos = linea.split()

            # Validemos que cada línea tiene 6 columnas
            if len(campos) != 6:
                print(f"Advertencia: Línea {i} ignorada - '{linea}', no tiene 6 columnas (tiene {len(campos)})")
                lineas_con_error += 1
                continue  # Saltar esta línea y continuar con la siguiente

            # Convertir cada campo (6 en cada fila) en un número float
            try:
                fila = [float(campo) for campo in campos]
                datos.append(fila)
            except ValueError as e:
                print(f"Advertencia: Línea {i} ignorada - '{linea}' - valor no numérico: {e}")
                lineas_con_error += 1
                continue  # Saltar esta línea y continuar con la siguiente

        if lineas_con_error > 0:
            print(f"\nSe ignoraron {lineas_con_error} líneas con errores")
            print(f"Se procesaron correctamente {len(datos)} líneas")
        else:
            print(f"Todas las {len(datos)} líneas se procesaron correctamente")

        return datos

    # Bloque except1
    except FileNotFoundError:
        print("Error, el archivo no se puede cargar")
        return []
    except Exception as e:
        print(f"Error inesperado: {e}")
        return []
```

```python
# Paso 1: Filtrar datos del distrito "77th Street"
calle77 = crime_la.loc[
    crime_la["area_name"] == "77th Street",
    ["lat", "lon"]
].dropna()

# Verificar que hay datos
if calle77.empty:
    raise ValueError("No hay datos para el distrito '77th Street'")
print(f"Se encontraron {len(calle77)} crímenes con coordenadas válidas en el distrito '77th Street'")

# Paso 2: Calculamos el centro del mapa (Centroide de los puntos calle77)
center_lat = calle77["lat"].mean()
center_lon = calle77["lon"].mean()

# Paso 3: Crear mapa de folium
m77 = folium.Map(location=[center_lat, center_lon], zoom_start=12)

# Agregar los puntos (lat, lon) de los crímenes


# Agreguemos un marcador en el centro de la calle77
folium.Marker(
    location=[center_lat, center_lon],
    popup="Centro de la calle 77",
    icon=folium.Icon(color="red", icon="info-sign")
).add_to(m77)

# Agregar los puntos donde ocurrió un crimen
for idx, row in calle77.iterrows():
    folium.CircleMarker(
        location=[row["lat"], row["lon"]],
        radius=2, # Tamaño más pequeño que el valor por defecto
        color="blue",
        fill=True,
        fillColor = "red"
        ).add_to(m77)

# Mostrar el mapa
m77
```

```python
# Seleccionemos a: Hollywood, Wilshire

# Paso 1: Definir áreas y colores
areas_colores = {
    "Hollywood": "blue",
    "Wilshire": "green"
}

# Filtrar y combinar los datos
puntos = []
for area, color in areas_colores.items():
    df_area = crime_la.loc[
        crime_la["area_name"] == area,
        ["lat", "lon"]
    ].dropna()

    # Añadir color y nombre al dataframe
    df_area = df_area.assign(area = area, color = color)
    puntos.append(df_area)

# Combinamos en un solo dataframe
puntos = pd.concat(puntos, ignore_index=True)

# Recomendación: Seleccionamos una muestra (20% del total)
puntos_df = puntos.sample(frac=0.2)

# Creamos el mapa
m_h_w = folium.Map(location = [puntos_df["lat"].mean(), puntos_df["lon"].mean()],
                   zoom_start = 12)

# Agregar los puntos (lat, lon) de los crímenes
for area, color in areas_colores.items():
    df = puntos_df[puntos_df["area"] == area]
    for _, row in df.iterrows():
        folium.CircleMarker(
            location = [row["lat"], row["lon"]],
            radius = 2,
            color = color,
            fill = True,
            fillColor = color,
            popup = f"Crimen en {area}"
        ).add_to(m_h_w)

# Mostremos el mapa
m_h_w
```

```python
# Necesito validad la existencia del archivo a descartar. En caso de que exista
# no deberé descarcarlo nuevamente. Luego también debo descomprimirlo

import os
import zipfile
import urllib.request

url = "https://github.com/robintux/Datasets4StackOverFlowQuestions/raw/refs/heads/master/cb_2018_06_bg_500k.zip"
zip_filename = "cb_2018_06_bg_500k.zip"

# Nombre esperado del shp principal (sin extensión)
shapefile_base = "cb_2018_06_bg_500k"
required_files = [f"{shapefile_base}{ext}" for ext in [".shp", ".shx", ".dbf", ".prj"]]

# Paso 01: Verificación si los archivos ya están descomprimidos
if all(os.path.exists(f) for f in required_files):
    print("Los archivos ya están descomprimidos. No se requieren descompresión.")
else:
    # Paso 02: Verificar si el zip existe
    if not os.path.exists(zip_filename):
        print(f"Descargando {zip_filename}...")
        urllib.request.urlretrieve(url, zip_filename)
        print(f"Archivo descargado en {os.path.abspath(zip_filename)}")
    else:
        print(f"El archivo {zip_filename} ya existe")

    # Paso 03: Descomprimir el zip
    print(f"Descomprimiendo {zip_filename}...")
    with zipfile.ZipFile(zip_filename, "r") as zip_ref:
        zip_ref.extractall(".")
    print(f"Archivo descomprimido en {os.getcwd()}")
```

```python
# Carguemos ese archivo de extensión shp
import geopandas as gpd
gdf = gpd.read_file("cb_2018_06_bg_500k.shp")

# Mostrar las primeras 5 filas
gdf.head(5)

# Mostrar las últimas 5 filas
gdf.tail(5)
```

```python
# Cálculo del centroide para centrar el mapa
centroid = gdf.geometry.union_all().centroid
center_lat, center_lon = centroid.y, centroid.x

# Creamos el mapa base
m_california = folium.Map(
    location = [center_lat, center_lon],
    zoom_start = 6
)

# Añadir los polígonos del mapa
folium.GeoJson(
    gdf.__geo_interface__,
    name = "California 2018",
    style_function = lambda x: {
        "fillColor": "blue",
        "color": "lightblue",
        "weight": 0.5,
        "fillOpacity": 0.2
    },
    tooltip = folium.GeoJsonTooltip(
        fields = ["GEOID", "NAME"],
        aliases = ["ID Geográfico", "Nombre"],
        localize = True,
        sticky = False,
        labels = True
        )
).add_to(m_california)


# Mostremos el mapa
m_california
```
